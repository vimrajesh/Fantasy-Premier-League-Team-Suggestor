```python


def onpick(event):
    
    name = df.iloc[event.ind[0]]['name']
    roi  = df.iloc[event.ind[0]]["ROI"]
    tota = df.iloc[event.ind[0]]["total_points"]
    plt.gca().set_title('{} ({},{})'.format(name,tota,roi))
    
# tell mpl_connect we want to pass a 'pick_event' into onpick when the event is detected
plt.gcf().canvas.mpl_connect('pick_event', onpick)
```

```python
def select_players(gk = 2, defe = 5, mid =5,fwd =0, moneytotal = 80.5):
    exp_players = 2
    total_points = 0
    roi = 0
    fdr_init = [
2.666666667,
2.666666667,
2.666666667,
2.666666667,
3.333333333,
2.666666667,
4,
3.666666667,
2.666666667,
2.666666667,
3,
3.333333333,
3.666666667,
3,
3,
3.333333333,
3.333333333,
3.333333333,
2.333333333,
3.333333333
]


    positions = { "GKP" : gk , "MID" : mid , "DEF" : defe, "FWD" : fwd}
    strteam = "ARS AVL BHA BUR CHE CRY EVE FUL LEE LEI LIV MCI MUN NEW SHU SOU TOT WBA WHU WOL"
    init_diff= {}
    strteam= strteam.split()
    team_limit = {}
    for i in range(len(strteam)):
        init_diff[strteam[i]] = fdr_init[i]
    for item in strteam:
        team_limit[item] = 3 #2
#     team_limit["LIV"] = 3
#     team_limit["SOU"] = 1
    # teams = { "" : }
    final_team = []
    count_players = 0
    roi_players = 10
    # print(moneytotal)
    cummulative_cost = 0
    liscost = []
    names_only = []
    #players i dont want
    details = [("Henderson","MUN") , ("Aurier","TOT"),("James","MUN"), ("de Gea","MUN"),("Fernández","TOT"),("Shelvey","NEW")
              ,('Dendoncker',"WOL")]
    # Taking star players
    for idx,r in knapsackdf1.iterrows():
        if exp_players==0:
            break
        else:
            nm = r["name"]
            clb = r["team"]
            pos = r["position"]
            
            if (nm,clb) in details:
                continue
            if clb not in strteam:
                continue
            if init_diff[clb]<=2.7:
                continue
            if nm in gw_initial.values and (nm,clb) not in names_only:
                if nm not in injured_list_initial.values or ((len(injured_list_initial[injured_list_initial["Name"] == nm] > 0)
                                                             and (injured_list_initial[injured_list_initial["Name"] == nm]["Chance to Play"].values > 70))):
                    
                    cst =( gw_initial[(gw_initial["second_name"] == nm )|( gw_initial["first_name"] == nm )]["now_cost"].values)[0]
                    
                    if(positions[pos]) > 0 and (team_limit[clb] > 0 and moneytotal>=cst):
                        names_only.append((nm,clb))
                        total_points = total_points + r["total_points"]
                        final_team.append((nm,pos,cst,clb,r["total_points"],r["ROI"]))
                        roi =  roi + r["ROI"]
                        moneytotal = moneytotal - cst
                        exp_players = exp_players -1
                        positions[pos] = positions[pos] - 1
                        team_limit[clb] = team_limit[clb] -1
                        cummulative_cost = cummulative_cost + cst 
                        liscost.append(cummulative_cost)

    for idx,r in knapsackdf2.iterrows():
        if roi_players < 0:
            break;
        else:
            nm = r["name"]
            clb = r["team"]
            pos = r["position"]
            
            if clb not in strteam:
                continue
            if (nm,clb) in details:
                continue
            if init_diff[clb]<=2.7:
                continue
            if nm in gw_initial.values and (nm,clb) not in names_only:
                if nm not in injured_list_initial.values or ((len(injured_list_initial[injured_list_initial["Name"] == nm]) > 0)
                                                             and ((injured_list_initial[injured_list_initial["Name"] == nm]["Chance to Play"].values) > 70)):
                   
                    cst =( gw_initial[(gw_initial["second_name"] == nm )|( gw_initial["first_name"] == nm )]["now_cost"].values)[0]
                    if(nm == "Long" and clb == "SOU"):
                        cst = 5.5
                    curcst = moneytotal - cst
                    curplay = (roi_players -1) * 4.9
                    if(positions[pos]) > 0 and (team_limit[clb] > 0) and (moneytotal>=cst) and (curcst >= curplay):
                        names_only.append((nm,clb))
                        total_points = total_points + r["total_points"]
                        final_team.append((nm,pos,cst,clb,r["total_points"],r["ROI"]))
                        roi =  roi + r["ROI"]
                        moneytotal = moneytotal - cst
                        roi_players = roi_players -1
                        positions[pos] = positions[pos] - 1
                        team_limit[clb] = team_limit[clb] -1
                        cummulative_cost = cummulative_cost + cst 
                        liscost.append(cummulative_cost)
#     print(len(final_team))
#     print(final_team)
#     print("last year ", total_points)
#     print(moneytotal)
#     print(roi)

    cst =( gw_initial[(gw_initial["second_name"] == "Wilson" )&( gw_initial["first_name"] == "Callum" )]["now_cost"].values)[0]
    str1 = knapsackdf2[knapsackdf2["name"] == "Callum Wilson"].values.tolist()
    final_team.append((str1[0][0],str1[0][2],cst,str1[0][1],str1[0][4],str1[0][7]))
    cummulative_cost = cummulative_cost + cst 
    liscost.append(cummulative_cost)
    # print(cummulative_cost)
    # print(str1)
    
    cst =( gw_initial[(gw_initial["second_name"] == "Adams" )&( gw_initial["first_name"] == "Che" )]["now_cost"].values)[0]
    str2= knapsackdf2[(knapsackdf2["name"] == "Adams") & (knapsackdf2["position"] == "FWD")].values.tolist()
    final_team.append((str2[0][0],str2[0][2],cst,str2[0][1],str2[0][4],str2[0][7]))
    cummulative_cost = cummulative_cost + cst 
    liscost.append(cummulative_cost)
    
    cst =( gw_initial[(gw_initial["second_name"] == "Calvert-Lewin" )|( gw_initial["first_name"] == "Calvert-Lewin" )]["now_cost"].values)[0]
    str3= knapsackdf2[(knapsackdf2["name"] == "Calvert-Lewin") & (knapsackdf2["position"] == "FWD")].values.tolist()
    final_team.append((str3[0][0],str3[0][2],cst,str3[0][1],str3[0][4],str3[0][7]))    
    cummulative_cost = cummulative_cost + cst 
    liscost.append(cummulative_cost)
    
    print(liscost)
    print(positions)
    
    return final_team
select_players()        
```
