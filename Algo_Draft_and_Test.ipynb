```python


def onpick(event):
    
    name = df.iloc[event.ind[0]]['name']
    roi  = df.iloc[event.ind[0]]["ROI"]
    tota = df.iloc[event.ind[0]]["total_points"]
    plt.gca().set_title('{} ({},{})'.format(name,tota,roi))
    
# tell mpl_connect we want to pass a 'pick_event' into onpick when the event is detected
plt.gcf().canvas.mpl_connect('pick_event', onpick)
```

```python
knapsackdf1 = df.copy(deep=True)
knapsackdf1
```

```python
knapsackdf1= knapsackdf1.sort_values(by="total_points",ascending= False)
knapsackdf1
```

```python
knapsackdf2= knapsackdf2.sort_values(by="ROI",ascending= False)
knapsackdf2
```

```python
gw_initial  = pd.read_csv("cleaned_players.csv")
gw_initial["now_cost"] = gw_initial["now_cost"] / 10
gw_initial = gw_initial[["first_name","second_name","now_cost"]]
gw_initial
```

```python
injured_list_initial = pd.read_excel('Injured list GW1.xlsx')
injured_list_initial.drop("id",axis = 1,inplace = True)
injured_list_initial = injured_list_initial[["Name","Team","Position","Availability","Chance to Play"]]
injured_list_initial.fillna(0,inplace=True)
injured_list_initial
```

# Constraints 
1. Maximum 3 Players per team
2. Number of Players by respective position :-
a) GK - 2
b) DEF - 5
c) MID - 5
a) FWD - 3
3. And the main one
BUDGET - $100 M

```python
def select_players(gk = 2, defe = 5, mid =5,fwd =3, moneytotal = 100):
    exp_players = 2
    total_points = 0
    roi = 0
    positions = { "GKP" : gk , "MID" : mid , "DEF" : defe, "FWD" : fwd}
    strteam = "ARS AVL BHA BUR CHE CRY EVE FUL LEI LEE LIV MCI MUN NEW SHU SOU TOT WBA WHU WOL"
    strteam= strteam.split()
    team_limit = {}
    for item in strteam:
        team_limit[item] = 3
    # teams = { "" : }
    final_team = []
    count_players = 0
    roi_players = 13
    print(moneytotal)
    cummulative_cost = 0
    liscost = []
    # Taking star players
    for idx,r in knapsackdf1.iterrows():
        if exp_players==0:
            break
        else:
            nm = r["name"]
            clb = r["team"]
            pos = r["position"]
            
            if clb not in strteam:
                continue
            if nm in gw_initial.values:
                if nm not in injured_list_initial.values or ((len(injured_list_initial[injured_list_initial["Name"] == nm] > 0)
                                                             and (injured_list_initial[injured_list_initial["Name"] == nm]["Chance to Play"].values > 70))):
                    cst =( gw_initial[(gw_initial["second_name"] == nm )|( gw_initial["first_name"] == nm )]["now_cost"].values)[0]

                    if(positions[pos]) > 0 and (team_limit[clb] > 0 and moneytotal>=cst):
                        total_points = total_points + r["total_points"]
                        final_team.append(nm)
                        roi =  roi + r["ROI"]
                        moneytotal = moneytotal - cst
                        exp_players = exp_players -1
                        positions[pos] = positions[pos] - 1
                        team_limit[clb] = team_limit[clb] -1
                        cummulative_cost = cummulative_cost + cst 
                        liscost.append(cummulative_cost)

    for idx,r in knapsackdf2.iterrows():
        if roi_players == 0:
            break;
        else:
            nm = r["name"]
            clb = r["team"]
            pos = r["position"]
            if clb not in strteam:
                continue
            if nm in gw_initial.values:
                if nm not in injured_list_initial.values or ((len(injured_list_initial[injured_list_initial["Name"] == nm]) > 0)
                                                             and ((injured_list_initial[injured_list_initial["Name"] == nm]["Chance to Play"].values) > 70)):
                    cst =( gw_initial[(gw_initial["second_name"] == nm )|( gw_initial["first_name"] == nm )]["now_cost"].values)[0]
                    curcst = moneytotal - cst
                    curplay = (exp_players -1) * 4
                    if(positions[pos]) > 0 and (team_limit[clb] > 0) and (moneytotal>=cst) and (curcst >= curplay):
                        total_points = total_points + r["total_points"]
                        final_team.append(nm)
                        roi =  roi + r["ROI"]
                        moneytotal = moneytotal - cst
                        exp_players = exp_players -1
                        positions[pos] = positions[pos] - 1
                        team_limit[clb] = team_limit[clb] -1
                        cummulative_cost = cummulative_cost + cst 
                        liscost.append(cummulative_cost)
    print(final_team)
    print(total_points)
    print(moneytotal)
    print(liscost)
    print(positions)
select_players()        
```
